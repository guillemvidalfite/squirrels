**3)**  The probability of a pet being a dog is 30%. The probability that a pet dog weighs more than 3 kg is 60%. If a pet is not a dog, the probability of it weighing less than or equal to 3 kg is 70%. What is the probability of a pet being a dog if it weighs more than 3 kg?

**Answer**:

The probability of a pet being a dog is 30%:\
*P(dog) = 0.3*

The probability that a pet dog weighs more than 3 kg is 60%:\
*P(3k/dog) = 0.6*

If a pet is not a dog, the probability of it weighing less than or equal to 3 kg is 70%:\
*P(3k'/dog') = 0.7*

Question: *P(dog/3k) = ?*

According to the bayes formula: *P(dog/3k)=(P(3k/dog).P(dog)) / P(3k)*

We do not have P(3k) (the probability of a pet weighting >3kg)

Assuming *P(3k) = P(3k/dog) + P(3k/dog')* \
and as *P(3k/dog') = 1 - P(3k'/dog')* \
then *P(3k) = P(3k/dog) + 1 - P(3k'/dog')*

*P(dog/3k) = (P(3k/dog).P(dog)) / (P(3k/dog) + 1 - P(3k'/dog')) = (0.6 * 0.3) / (0.6 + 1 - 0.3) = 0.9 / 1.3 = 0.69*

The resulting probability is **0.69**

### Data clean

```python
# remove columns we're not going to use
squirrels.drop(columns=["Color notes", 
                        "Specific Location", 
                        "Other Activities",  
                        "Lat/Long", 
                        "Highlight Fur Color",
                        "Age"], inplace=True)

squirrels.shape
```

```python
# extract day from date (month and year are all the same)
squirrels["day"] = squirrels["Date"].astype(str).str.slice(2, 4)

# extract hectare number and letter for further plots
squirrels["hectare_num"] = squirrels["Hectare"].astype(str).str.slice(0, 2)
squirrels["hectare_letter"] = squirrels["Hectare"].astype(str).str.slice(2, 3)

# above groud feet make numeric. False (ground measurements can be set to 0)
squirrels["above_ground_feet"] = np.where(squirrels["Above Ground Sighter Measurement"] == "FALSE", 0, squirrels["Above Ground Sighter Measurement"])
squirrels["on_the_ground"] = np.where(squirrels["Above Ground Sighter Measurement"] == "FALSE", 1, 0)

# remove useless columns
squirrels.drop(columns=["Date", "Above Ground Sighter Measurement", "Hectare", "Location"], inplace=True)
squirrels.shape
  
```

## Events geolocation

```python
squirrel_events = squirrels[["X", "Y",  "Running", "Chasing", "Climbing", "Eating", "Foraging", "Tail flags", "Tail twitches", "Kuks", "Quaas", "Moans", "Approaches", "Indifferent", "Runs from"]]
squirrel_events.shape
```

```python
# Procedure to plot an event geographically
def plot_geo_event(df_events, event_name):
    geo_df = df_events[["X", "Y", event_name]]
    df_true = geo_df[geo_df[event_name] == 1]
    df_false = geo_df[geo_df[event_name] == 0]
    
    ax = df_true.plot(x="X", y="Y", kind="scatter", c='r', label=event_name, 
                  xlabel="X_coord", ylabel="Y_coord", title=event_name + " observations",
                  figsize=(13,13))

    df_false.plot(x="X", y="Y", kind="scatter", ax=ax, c='b', label="Non "+event_name)
    
    return ax
```

```python
plot_geo_event(squirrel_events,"Running")
```

```python
plot_geo_event(squirrel_events,"Climbing")
```

```python
plot_geo_event(squirrel_events,"Foraging")
```

After observing squirrel interaction events geographically, all events seem to appear spreaded across the Park. No specific event locations appear at this level.
